A. Lakukan operasi arimatika menggunakan gambar 1 dan 2, sehingga menghasilkan output gambar berikut
image.png

```python
# kolom jawaban
dest_not2 = cv2.bitwise_not(Image_size1, mask = None)
dest_not3 = cv2.bitwise_not(Image_size2, mask = None)
dest_and = cv2.bitwise_and(dest_not2, dest_not3, mask = None)
dest_not3 = cv2.bitwise_not(dest_and , mask = None)
plt.imshow(dest_not3)
plt.show()
```

# Soal operasi geometri (20 points)
Instruksi : membuat code python

Lakukan operasi geometri pada gambar, sehingga menghasilkan output gambar berikut

Hint : translasi (pergeseran) gambar sejauh 50 piksel ke kanan dan 50 piksel ke bawah.
image.png

```python
# kolom jawaban
image_rgb = cv2.cvtColor(img_4 , cv2.COLOR_BGR2RGB)

width = image_rgb.shape[1]
height = image_rgb.shape[0]

tx = 50
ty = 50

# Translation matrix
translation_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=np.float32)
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (width, height))

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

# Plot the original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

# Plot the transalted image
axs[1].imshow(translated_image)
axs[1].set_title('Image Translation')

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```

## Soal 4 (30 points)
Soal: Kombinasikan Operasi Geometri, Aritmatika, dan Logika pada Gambar
Gunakan dua gambar sebelumnya (gambar_2.png dan gambar_3.png). Tulislah program Python menggunakan OpenCV dan NumPy untuk melakukan langkah-langkah berikut:

1. Operasi Geometri:


*   Lakukan rotasi pada gambar_2.png sebesar 30 derajat searah jarum jam dengan pusat rotasi berada di tengah gambar.
*   Lakukan translasi (pergeseran) pada gambar_3.png dengan menggeser gambar sejauh 10 piksel ke kanan dan 50 piksel ke bawah.


2. Operasi Aritmatika:
*   Gabungkan kedua gambar hasil operasi geometri menggunakan penjumlahan (cv2.add).

3. Operasi Logika:
*   Lakukan operasi NOT (cv2.bitwise_not) pada hasil penjumlahan untuk menghasilkan gambar akhir.



**Instruksi:**

Pastikan kedua gambar memiliki dimensi yang sama sebelum melakukan operasi.
Tampilkan hasil dari setiap langkah operasi: hasil rotasi, hasil translasi, hasil penjumlahan, dan hasil akhir setelah operasi NOT.

Gunakan fungsi matplotlib.pyplot untuk menampilkan gambar hasil setiap tahap dengan menggunakan plt.imshow().

```python
# Convert BGR image to RGB
image_rgb = cv2.cvtColor(Image_size2, cv2.COLOR_BGR2RGB)
image_rgb1= cv2.cvtColor(Image_size3, cv2.COLOR_BGR2RGB)

# Image rotation parameter
width = image_rgb.shape[1]
height = image_rgb.shape[0]

center = (image_rgb.shape[1] // 2, image_rgb.shape[0] // 2)
center = (image_rgb1.shape[1] // 2, image_rgb1.shape[0] // 2)
angle = -30
scale = 1

tx = 10
ty = 50

# getRotationMatrix2D creates a matrix needed for transformation.
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# We want matrix for rotation w.r.t center to 30 degree without scaling.
rotated_ima2 = cv2.warpAffine(image_rgb, rotation_matrix, (Image_size2.shape[1], Image_size2.shape[0]))
rotated_ima1= cv2.warpAffine(image_rgb1, rotation_matrix, (Image_size3.shape[1], Image_size3.shape[0]))
translation_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=np.float32)
translated_image = cv2.warpAffine(image_rgb1, translation_matrix, (width, height))
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title("Original")

plt.subplot(2, 3, 2)
plt.imshow(rotated_ima2)
plt.title("Image Rotation 1")

plt.subplot(2, 3, 3)
plt.imshow(image_rgb1)
plt.title("Original 2")

plt.subplot(2, 3, 4)
plt.imshow(rotated_ima1)
plt.title("Image Rotation 2")

plt.subplot(2, 3, 5)
plt.imshow(image_rgb1)
plt.title("Original")

plt.subplot(2, 3, 6)
plt.imshow(translated_image)
plt.title("pergeseran")

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```
